```python
df = pd.read_csv('/content/Sales Dataset.csv')
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
df.dropna(inplace=True)  # Example of removing missing values
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')

#Converting any date columns to datetime objects
```

```python
df.drop_duplicates(inplace=True)
```

# D**ata Analysis**

Identify top-performing products:

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

Analyze seasonal sales patterns:

```python
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')['Sales'].sum()
print(monthly_sales)
```

Visualize the sales trends:

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales)
plt.title('Monthly Sales Trend')
plt.show()
```

Identify sales trends over time

```python
sales_trends = df.groupby('Order Date')['Sales'].sum()
print(sales_trends)
```

# **Visualization**

Plot top-performing products

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=top_products.index, y=top_products.values)
plt.xticks(rotation=45)
plt.title('Top Performing Products')
plt.show()
```

Plot seasonal sales patterns

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Seasonal Sales Patterns')
plt.show()
```
